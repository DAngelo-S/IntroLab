```python
import csv
import requests
import matplotlib.pyplot as plt
```

```python
def get_data(name):
  url = f"https://raw.githubusercontent.com/DAngelo-S/IntroLab/refs/heads/main/{name}"
  response = requests.get(url)

  dados = response.content.decode('utf-8').splitlines()

  reader = csv.reader(dados, delimiter=';')

  t, z, v, a = [], [], [], []

  i = 0
  n = sum(1 for line in reader)
  reader = csv.reader(dados, delimiter=';')

  for line in reader:
    if i < 2:
      i += 1
      continue
    t.append(float(line[0].replace(',','.')))
    z.append(float(line[1].replace(',','.')))
    if i < 3 or i >= n - 1:
      i += 1
      continue
    v.append(float(line[2].replace(',','.')))
    if i < 4 or i >= n - 2:
      i += 1
      continue
    a.append(float(line[3].replace(',','.')))
    i += 1

  return t, z, v, a
```

# Grafo 6

```python
t, z, v, a = get_data("dados06.txt")

t = [i - min(t) for i in t]
t = [i for i in t if i > 3]
t = [i - min(t) for i in t]

n = len(t)

z = z[-n:]
v = v[-n:]
a = a[-n:]
```
